# Importing the libraries and reading the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl_df = pd.read_csv('matches.csv')
ipl_df.head()
```

```python
ipl_df.info()
```

# Picking records for the year 2017 by using the column 'date' even though we can pick on the column 'season'

```python
ipl_df[(ipl_df.date.str.contains('/2017', regex=True, na=False)) | (ipl_df.date.str.contains('/17', regex=True, na=False))]
```

# Listing the Winners of each season

```python
ipl_winners = ipl_df[['season','team1','team2','winner']].groupby('season').tail(1).sort_values(by='season')
ipl_winners
```

# Plot showing the number of IPL Season wins for each team

```python
fig_dims=(18,5)
fig, new_ax = plt.subplots(figsize=fig_dims)
sns.countplot(ipl_winners['winner'])
```

# Plot showing the number of matches won by each team for 2017 season

```python
fig_dims=(18,5)
fig, new_ax = plt.subplots(figsize=fig_dims)
sns.countplot(ipl_df[ipl_df['season']==2017]['winner'])
```

# Determining the ratio of matches won over the number of tosses won for each team

```python
ipl_tw = ipl_df[['toss_winner']].groupby(ipl_df['toss_winner']).count()
ipl_tw = ipl_tw.rename(columns={'toss_winner':'toss_win_count'})
ipl_tw.reset_index(inplace=True)
ipl_tw
```

```python
tw_eq_win = ipl_df[ipl_df['toss_winner']==ipl_df['winner']][['winner']].groupby(ipl_df['winner']).count()
tw_eq_win = tw_eq_win.rename(columns={'winner':'win_count'})
tw_eq_win.reset_index(inplace=True)
tw_eq_win['toss_win_count'] = ipl_tw['toss_win_count']
tw_eq_win['win_per_tosswin'] = tw_eq_win['win_count']/tw_eq_win['toss_win_count']*100
tw_eq_win
```

```python
fig_dims=(36,12)
fig, new_ax = plt.subplots(figsize=fig_dims)
sns.barplot(tw_eq_win['winner'],tw_eq_win['win_per_tosswin'])
```

```python
fig_dims=(36,12)
fig, new_ax = plt.subplots(figsize=fig_dims)
sns.pointplot(tw_eq_win['winner'],tw_eq_win['win_per_tosswin'],color='red')
```

# Determining the count of matches won while fielding first and batting first when a team won both the toss and match

```python
td_df = ipl_df[ipl_df['toss_winner']==ipl_df['winner']][['toss_winner','winner','toss_decision']]
```

```python
td_df.groupby(['winner','toss_decision']).count()
```

# Plot showing matches won while fielding first and batting first.

The Plot shows that almost every team won matches fielding first when they won the toss. Chennai Super Kings stand out to be an exception.

```python
fig_dims=(36,12)
fig, new_ax = plt.subplots(figsize=fig_dims)
sns.countplot(td_df['winner'],hue=td_df['toss_decision'],palette='bright')
```
